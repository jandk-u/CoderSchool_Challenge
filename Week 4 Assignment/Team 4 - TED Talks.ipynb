```python
import numpy as np
import pandas as pd
import seaborn as sns
import ast
```

```python
"""
Trong data chứa các thông tin về
Comment : Bình luận 
Description: Mô tả
Duration : Thời lượng
Event : Tên sự kiện 
Film date: Ngày quay film (đơn vị giây)
Languages: số lượng ngôn ngữ sử dụng
Main speaker: Người nói chính
Published date: Ngày xuất bản
Rating: Xếp hạng( 1 dictionary bao gồm các thông tin về id, name, count)
Related talks: Những cuộc nói chuyện liên quan bao gồm id, hero
Speaker occupation: Nghề nghiệp của người nói
Tags: Nhãn
Title: Tiêu đề
Url: Đường dẫn
Views: Số lượt xem
"""
```

```python
# Data Ratings
```

dataTed['ratings']

```python
# Lấy các categories trong rating 
rates = dataTed['ratings']

rating_name = set()
for row in rates:
    rating = ast.literal_eval(row)
    for item in rating:
        rating_name.add(item['name'])
rating_name
```

```python
#Ratings data
"""
Chuyển đổi rating về thành một DataFrame 
"""
from collections import  defaultdict

dataTed['Talk_id'] = range(1, len(dataTed)+1)
rating_data = defaultdict(list)

for index, row in dataTed.iterrows():
    rating = ast.literal_eval(row['ratings'])
    rating_data['Talk_id'].append(row['Talk_id'])
    names = set()
    for item in rating:
        rating_data[item['name']].append(item['count'])
        names.add(item['name'])
rating_data = pd.DataFrame(rating_data)
rating_data.head()
```

```python
rating_data['Total'] = rating_data.sum(axis=1)
rating_data = rating_data.sort_values('Total', ascending=False)
```

```python
def column_percentage(column):
    return (column/rating_data['Total'])*100
rating_data.loc[:, (rating_data.columns!= 'Total')& (rating_data.columns!= 'Talk_id')] = rating_data.loc[:, (rating_data.columns!= 'Total')& (rating_data.columns!= 'Talk_id')].apply(column_percentage)

rating_data.head()
```

```python
count_talks = defaultdict(list)
for rating in rating_data.columns:
    if (rating != 'Talk_id') & (rating != 'Total'):
        count_talks['rating'].append(rating) 
        count_talks['count'].append(rating_data[rating_data[rating] >0][rating].count())
```

```python
import matplotlib.pyplot as plt
sns.barplot(x="rating", y="count", data=count_talks)
plt.ylim(2400, 2600)
plt.xticks(rotation='vertical')
```

```python
rating_values = rating_data.loc[:, (rating_data.columns != 'Total') &  (rating_data.columns !='Talk_id')] 
rating_data['main_rating'] = rating_values.apply(np.argmax, axis = 1)

rating_data['main_rating'].head()
```

```python
label_order = rating_data['main_rating'].value_counts().sort_values(ascending = False).index

f, ax = plt.subplots(figsize=(12, 6))
ax = sns.countplot(x = 'main_rating', data = rating_data, order = label_order)
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2., height + 3,'{:d}'.format(height),ha="center") 
plt.xticks(rotation=30)
```

```python
tags_data = defaultdict(list)

for index, row in dataTed.iterrows():
    tags = ast.literal_eval(row['tags'])

    for item in tags:
        tags_data['Talk_id'].append(row['Talk_id'])
        tags_data['tags'].append(item)
tags_data = pd.DataFrame(tags_data)
```

```python
# chuyển đổi ngày tháng
import datetime
def trainsform_date(date):
    return datetime.date.fromtimestamp(date)

dataTed['film_date'] = dataTed['film_date'].apply(trainsform_date)
dataTed['published_date'] = dataTed['published_date'].apply(trainsform_date)

dataTed['duration'] = dataTed['duration']/60
dataTed.head(5)
```

```python
month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
```
